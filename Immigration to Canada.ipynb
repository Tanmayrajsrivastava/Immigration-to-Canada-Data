## Data Cleaning

```python
df.head()
```

```python
# Dropping unnecessary data

# For rows, axis = 0 & for columns, axis = 1; by default rows are selected 


df.drop(['AREA','REG','DEV','Type','Coverage'], axis =1)
```

```python
# For dropping them permanently *inplace = True

df.drop(['AREA','REG','DEV','Type','Coverage'], axis =1, inplace = True)
df.head()
```

```python
# Column Renaming

df.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df.columns      
```

```python
# Column Addition

df['Total'] = df.sum(axis=1)
```

```python
# Checking the total of empty values present in the dataset

df.isnull().sum()
```

```python
# Describing the dataset

df.describe()
```

## Data Comparision

```python
# Here, taking data for INDIA and CHINA

df_CI = df.loc[['India', 'China'], years]
df_CI.head()
```

```python
df_CI.plot(kind='line')
```

```python
#Since df_CI is a dataframe with the country as the index and years as the columns, 
#we must first transpose the dataframe using transpose() method to swap the row and columns.

df_CI = df_CI.transpose()
df_CI.head()
```

```python
# Plotting the graph
df_CI.index = df_CI.index.map(int) 
df_CI.plot(kind='line')

plt.title('Immigration from China and India')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # need this line to show the updates made to the figure


```
